# **2. 로지스틱 회귀(Logistic Regression)**
* 둘 중의 하나를 결정하는 문제(이진 분류)를 풀기 위한 대표적인 알고리즘
* 입력 데이터와 가중치의 선형 조합으로 선형 방정식을 만듦 -> 선형 방적식의 결과를 0과 1사이의 확률값으로 변환(시그모이드 함수)
* 3개 이상의 클래스에 대한 판별도 할 수 있음
    * OvR(One-vs-Rest): 각 클래스마다 하나의 이진 분류기를 만들고, 해당 클래스를 기준으로 그 클래스와 나머지 모든 클래스를 구분하는 이진 분류를 실행 -> 가장 높은 확률을 가진 클래스를 선택
    * OvO(One-vs-One): 클래스의 개수가 N인 경우 N(N-1)/2개의 이진 분류기를 만듦 -> 입력 데이터를 각 이진 분류기에 넣어 가장 많이 선택된 클래스를 최종 선택

> 대부분 OvR 전략을 선호. 클래스 간의 구분이 명확하지 않거나 데이터가 한쪽으로 치우친 경우 OvO를 고려


````python
from sklearn. linear_model import LogisticRegression
````

````python
lr=LogisticRegression() # 객체 만듬
````

````python
lr.fit(X_train ,y_train) # 학습
````

````python
from sklearn.metrics import accuracy_score
````

# **3. 혼돈 행렬(confusion matrix)**
* 정밀도와 재현율(민감도)을 활용하여 평가용 지수

```
TN(8784)   FP(100)
FN(673)    TP(175)
```

* TN: 승진하지 못했는데, 승진하지 못했다고 예측.
* FN: 승진했는데, 승진하지 못했다고 예측
* FP: 승진하지 못했는데, 승진했다고 예측.
* TP: 승진했는데, 승진했다고 예측.


````python
from sklearn.metrics import confusion_matrix
````

````python
confusion_matrix(y_test, pred)
````

````python
# seaborn 라이브러리의 heatmap 함수는 데이터를 히트맵 형태로 시각화하는 데 사용
# 혼동 행렬을 히트맵으로 시각화하면 각 클래스 간의 예측 정확도를 직관적으로 파악
sns.heatmap(confusion_matrix(y_test, pred), annot=True, cmap='Blues')
# annot=True: 각 셀에 값을 표시합니다.

# x축: 예측된 클래스 (Predicted Label)
# y축: 실제 클래스 (True Label)
````

### 3-1. 정밀도(precision)
* TP / (TP + FP)
* 무조건 양성으로 판단해서 계산하는 방법
* 실제 1인 것 중에서 얼마 만큼을 제대로 맞췄는가?

### 3-2. 재현율(recall)
* TP / (TP + FN)
* 정확하게 감지한 양성 샘플의 비율
* 1이라고 예측한 것 중, 얼마 만큼을 제대로 맞췄는가?
* 민감도 또는 TPR(True Positive Rate)라고도 부름

### 3-3. f1 score
* 정밀도와 재현율의 조화평균을 나타내는 지표

```
정밀도  재현율  산술평균  조화평균
0.4     0.6     0.5       0.48
0.3     0.7     0.5       0.42
0.5     0.5     0.5       0.5
```

$$2*\frac{정밀도 * 재현율}{정밀도 + 재현율}=\frac{TP}{TP+\frac{FN+FP}{2}}$$

* 산술평균: a+b/2
* 조화평균: 2ab/a+b

# **4. 교차 검증(Cross Validation)**
* train_test_split에서 발생하는 데이터의 섞임에 따라 성능이 좌우되는 문제를 해결하기 위한 방법
* K겹(K-Fold) 교차 검증을 가장 많이 사용

````python
# KFold는 scikit-learn에서 제공하는 교차 검증(cross-validation) 기법
from sklearn.model_selection import KFold
````

````python
# splits = 5 -> 데이터를 5개로 나눔  (훈련 4 테스트 1)
kf = KFold(n_splits=5)  # 몇 곂으로 나눌건지
kf
````

````python
hr_df
````

````python
# kf.split(range(len(hr_df)))는 KFold 객체를 사용하여
# 데이터프레임 hr_df의 인덱스를 기반으로 데이터를 K개의 fold로 나누는 과정을 나타냄.
# 각 반복(iteration)마다 train_index와 test_index로 나누어진 데이터의 인덱스를 출력하고,
# 각각의 길이도 함께 출력하는 코드
for train_index, test_index in kf.split(range(len(hr_df))):
    print(train_index, test_index, len(train_index), len(test_index))
# 번갈아 가면서 검증
# 섞임이 없음

# len(hr_df)는 데이터프레임의 전체 행 수
````

````python
# 섞는법
kf = KFold(n_splits=5, random_state=2024, shuffle=True)
kf
````

````python
for train_index, test_index in kf.split(range(len(hr_df))):
    print(train_index, test_index, len(train_index), len(test_index))
# 이제 섞임
# 5번 돌게 되어있음
````

````python
# KFold를 사용하여 위 데이터를 LogisticRefression 모델로 학습을 시키고
# 각 n마다 예측결과를 값으로 출력
````

````python
acc_list = []

for train_index, test_index in kf.split(range(len(hr_df))):  # 5바퀴 돔 #5개의 폴드로 나누겠다
    X = hr_df.drop('is_promoted', axis=1)
    y = hr_df['is_promoted']

# iloc은 Pandas 데이터프레임에서 특정 인덱스나 인덱스 범위를 사용하여 행을 선택할 때 사용하는 메소드
    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    y_train = y.iloc[train_index]
    y_test = y.iloc[test_index]

    lr = LogisticRegression()  # 객체만들고
    lr.fit(X_train, y_train)   # 학습
    pred = lr.predict(X_test)  # 예측
    acc_list.append(accuracy_score(y_test, pred))
````
