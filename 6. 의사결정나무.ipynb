```python
bike_df = bike_df.fillna(0)
bike_df.isna().sum()
```

```python
# pd.to_datetime(): Pandas 라이브러리의 to_datetime() 함수는 문자열 형식의 날짜 및 시간을 datetime 객체로 변환
bike_df['datetime'] = pd.to_datetime(bike_df['datetime'])
```

```python
# year, month, hour 파생변수 만들기
bike_df['year'] = bike_df['datetime'].dt.year  # dt는 pandas에서 날짜 및 시간 연산을 수행하기 위한 기능을 제공하는 속성
bike_df['month'] = bike_df['datetime'].dt.month
bike_df['hour'] = bike_df['datetime'].dt.hour
```

```python
bike_df.head()
```

```python
# dt는 Pandas에서 datetime 객체의 속성에 접근하기 위해 사용되는 접근자
bike_df['date'] = bike_df['datetime'].dt.date
```

```python
bike_df.head()
```

```python
# plt.figure(figsize=(14, 4)): 그래프의 크기를 설정
plt.figure(figsize=(14, 4))
# sns.lineplot(x='date', y='count', data=bike_df):
# Seaborn의 lineplot 함수를 사용하여 'date'를 x축으로, 'count'를 y축으로 하는 선 그래프를 그립니다.
sns.lineplot(x='date', y='count', data=bike_df)
# plt.xticks(rotation=45): x축의 날짜 레이블을 45도 회전
plt.xticks(rotation=45)
# 그래프를 화면에 출력
plt.show()
```

```python
# 2019년 월별 자전거 갯수를 출력
# bike_df 데이터프레임에서 연도가 2019년인 데이터를 필터링하고, 각 월별 평균 자전거 대여 횟수를 계산
bike_df[bike_df['year'] == 2019].groupby('month')['count'].mean() # mean(): 평균
```

```python
# 2020년 월별 자전거 갯수를 출력
bike_df[bike_df['year'] == 2020].groupby('month')['count'].mean()
# 2020년 4월 데이터가 없음을 알 수 있다
```

```python
# covid
# 2020-04-01 이전: precovid
# 2020-04-01 이후 ~ 2021-04-01 이전: covid
# 2021-04-01: postcovid
# 파생변수 covid

def covid(date):
    if str(date) < '2020-04-01':
        return 'precovid'
    elif str(date) < '2021-04-01':
        return 'covid'
    else:
        return 'postcovid'
```

```python
# .apply(covid): Pandas Series나 DataFrame의 요소에 함수를 적용하는 메서드
bike_df['date'].apply(covid)
```

```python
# lambda 이용
# lambda는 파이썬에서 익명 함수(anonymous function)를 생성하는 키워드입니다. 즉, 이름 없이 한 줄로 간단히 함수를 정의
bike_df['covid'] = bike_df['date'].apply(lambda date: 'precovid' if str(date) < '2020-04-01' else 'covid' if str(date)< '2021-04-01' else 'postcovid')
bike_df
```

```python
# season
# 12~2월: winter
# 3~5월: spring
# 6~8월: summer
# 9~11월: fall

bike_df['season'] = bike_df['month'].apply(lambda x: 'winter' if x == 12
                                           else 'fall' if x >= 9
                                           else 'summer' if x >= 6
                                           else 'spring' if x >= 3
                                           else 'winter')
bike_df[['month', 'season']]
```

```python
# day_night
# 21시 이후: night
# 19시 이후: late evening
# 17시 이후: early evening
# 15시 이후: late afternoon
# 13시 이후: early afternoon
# 11시 이후: late morning
# 6시 이후: early morning

bike_df['day_night'] = bike_df['hour'].apply(lambda x: 'night' if x >= 21
                                             else 'late evening' if x >= 19
                                             else 'early evening' if x >= 17
                                             else 'late afternoon' if x >= 15
                                             else 'early afternoon' if x >= 13
                                             else 'late morning' if x >= 11
                                             else 'early morning' if x >= 6
                                             else 'night')
bike_df.head()
```

```python
# 필요 없는 부분 날리기
bike_df.drop(['datetime', 'month', 'date', 'hour'], axis = 1, inplace=True)
bike_df.head()
```

```python
bike_df.info()
```

```python
# object 숫자로 바꾸기
for i in ['weather_main', 'covid', 'season', 'day_night']:
    print(i, bike_df[i].nunique()) # 갯수 => 다 살림
```

```python
bike_df['weather_main'].unique()  # 카테고리
```

```python
# pd.get_dummies() 함수는 카테고리형 열을 더미 변수로 변환하는 함수
bike_df = pd.get_dummies(bike_df, columns =['weather_main', 'covid', 'season', 'day_night'])
bike_df.head()
```

# **2. 의사 결정 나무 (Decision Tree)**
* 데이터를 분석하고 패턴을 파악하여 결정 규칙을 나무 구조로 나타낸 기계학습 알고리즘
* 간단하고 강력한 모델 중 하나로, 분류(Classification)과 회귀(Regression) 문제에 모두 사용
* 지니계수(지니 불순도, Gini Impurity): 분류 문제에서 특정 노드의 불순도를 나타내는데, 노드가 포함하는 클래스들이 혼잡되어 있는 정도를 나타냄
    * 0에서 1까지의 값을 가지며, 0에 가까울수록 노드의 값이 불순도가 없음을 의미
    * 로그 연산이 없어 계산이 상대적으로 빠름
* 엔트로피(Entropy): 어떤 집합이나 데이터릐 불확실성, 혼잡도를 나타내는데, 노드의 불순도를 측정하는데 활용
    * 0에서 무한대까지의 값을 가지며, 0에 가까울수록 노드의 값이 불순도가 없음을 의미
    * 로그 연산이 포함되어 있어 계산이 복잡
* 오버피팅(과적합): 학습데이터에서는 정확하나 테스트데이터에서는 성과가 나쁜 현상을 말합. 의사 결정 나무는 오버피팅이 매우 잘 일어남
    * 오버피팅을 방지하는 방법
        * 사전 가지치기: 나무가 다 자라기 전에 알고리즘을 멈추는 방법
        * 사후 가지치기: 의사결정 나무를 끝까지 돌린 후 밑에서부터 가지를 쳐나가는 방법


```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(random_state=2024)
```

```python
dtr.fit(X_train, y_train) # 학습시키기
```

```python
pred1 = dtr.predict(X_test) # 예측
pred1  # count가 나온것
```

```python
# 예측 시각화
sns.scatterplot(x=y_test, y=pred1) # 대칭이다 => 어느정도 예측값과 실제값이 일치
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# RMSE
mean_squared_error(y_test, pred1, squared=False)
```

# **3. 선형 회귀 vs 의사결정나무**

```python
from sklearn.linear_model import LinearRegression  # 예측 vs 예측
```

```python
lr = LinearRegression()
```

```python
# LinearRegression()에서 random_state를 사용하면 에러가 발생하는 것은 이 매개변수가 해당 모델에서 지원되지 않기 때문
lr.fit(X_train, y_train)  # 학습
```

```python
pred2 = lr.predict(X_test)
```

```python
# 예측 시각화
sns.scatterplot(x=y_test, y=pred2)
```

```python
mean_squared_error(y_test, pred2, squared=False)
```

```python
# 의사 결정 나무: 210.74186203651976
# 선형 회귀: 221.1987722244733
210.74186203651976 - 221.1987722244733
# 의사 결정 나무가 성능이 더 좋음
```

```python
# 하이퍼 파라미터 적용
# 하이퍼 파라미터 튜닝 => max_depth: 깊이 50까지, min_samples_leaf: 데이터가 쪼개지다가 30개 미만 되면 멈춰
dtr = DecisionTreeRegressor(random_state=2024, max_depth=50, min_samples_leaf=30)
```

```python
dtr.fit(X_train, y_train)  # 학습
```

```python
pred3 = dtr.predict(X_test)
```

```python
mean_squared_error(y_test, pred3, squared=False)
```

```python
# 의사 결정 나무 RMSE:  210.74186203651976
# 선형 회귀 RMSE:  221.1987722244733
# 의사 결정 나무 (하이퍼 파라미터 적용) RMSE: 181.29247853838177
181.29247853838177 - 210.74186203651976
# 파라미터 튜닝이 가장 성능이 좋음
```

```python
from sklearn.tree import plot_tree
```

```python
#  max_depth=5 위에서 5개만 보겠다.
plt.figure(figsize=(24, 12))
plot_tree(dtr, max_depth=5, fontsize=10)
plt.show()
```

```python
# feature_names=X_train.columns: 각 특성의 이름을 지정합니다. 이는 특성들이 어떤 열(column)을 나타내는지 명확히 표시
plt.figure(figsize=(24, 12))
plot_tree(dtr, max_depth=5, fontsize=10, feature_names=X_train.columns)
plt.show()
# 1 밤 낮
# 2 기온
```
